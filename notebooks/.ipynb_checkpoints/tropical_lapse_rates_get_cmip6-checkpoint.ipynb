Here we read CMIP6 data with intake, format the data a bit and calculate tropical means, precipitation weighted SSTs, precipitation weighted air temperatures. Also, the convecting regions are determined with precipitable water and the temperature in the convecting regions is written.

This script needs at least 64GB of RAM memory to compute and is constructed to run on the mistral machine from DKRZ. If you do not have access to mistral or a similar system, these types of data processing can also be done on the pangeo cloud, see here: https://pangeo.io/cloud.html

Things you might need to change:
- directory paths
- intake catalogue paths

It's recommended not to run the full script at once. Start with the first part (AMIP), which should be quicker.

If you want to check temperatures subsampled to radiosonde stations (not a Figure in the paper) you have to run the IUK processing script first.

All files will be stored in ../data.

# Define functions:

```python
def query(experiment, var) :
    query = dict(
        activity_id="CMIP",
        experiment_id=experiment,
        member_id="r1i1p1f1",
        variable_id=var,
        table_id="Amon",
        require_all_on = "source_id"
    )
    return query
```

```python
# clean data a bit before reading
def preproc(ds):
    if 'bnds' in ds.dims :
        ds = ds.drop_dims('bnds')
    if 'lat_bnds' in ds :
        ds = ds.drop('lat_bnds')
    if 'lon_bnds' in ds :
        ds = ds.drop('lon_bnds')
    if 'time_bounds' in ds :
        ds = ds.drop('time_bounds')
    if 'nbnd' in ds.dims :
        ds = ds.drop_dims('nbnd')
    return ds
```

```python
def calc_PRSST(ds,lat_bound) :
    
    # usually for an average only over SSTs, we need to convert the zeros to NaN
    # so they are not considered in the fieldmean
    # Here, because it is a ratio of averages, we do not need to account for this
    
    regridder = xe.Regridder(SST_mask, ds, 'bilinear', periodic=True, reuse_weights=True)
    mask_interpol = regridder(SST_mask)
    
    ds['SST'] = ds.ts * mask_interpol.sea
    ds_tropics = ds.sel(lat=slice(-lat_bound,lat_bound))
    
    PRSST = metcalc.xr_fldmean(ds_tropics.pr * ds_tropics.SST) / metcalc.xr_fldmean( ds_tropics.pr * mask_interpol.sea )
    
    return PRSST.to_dataset(name='PRSST')
    
```

```python
def calc_PRTA(ds_ts,ds_ta,lat_bound) :

    ds_ts_tropics = ds_ts.sel(lat=slice(-lat_bound,lat_bound))
    ds_ta_tropics = ds_ta.sel(lat=slice(-lat_bound,lat_bound))
    da_ta = ds_ta_tropics.sel(plev=70000).ta
    
    PRTA = metcalc.xr_fldmean(ds_ts_tropics.pr * da_ta) / metcalc.xr_fldmean( ds_ts_tropics.pr )
    
    return PRTA.to_dataset(name='PRTA')
```

## Process data

```python

def format_data(dset_dict, is_3D,lat_bound=20) :
    model_list = []
    ds_tm_ym_list=[]
    ds_list = []
    levels=[100000, 85000, 70000, 60000, 50000, 30000, 25000, 20000, 15000, 10000, 5000]
    all_levels = [100000., 92500., 85000., 70000., 60000., 50000., 40000., 30000., 25000.,
                20000.,  15000.,  10000., 7000., 5000., 3000., 2000., 1000., 500., 100.]
    pandas_time=pd.date_range(start='1/1979', end='01/2015', freq='M')



    for key, value in dset_dict.items():

        ds = dset_dict[key].squeeze()
        model = ds.source_id
        
        if is_3D == True :
            #for some models (ACCESS) the plev coordinate is weird. Therefore:
            ds['plev'] = all_levels
            ds = ds.sel(plev=levels)
        print("\n" + model )

        # one dataset has a 360 days calendar, so we go to december 30th:
        ds = ds.sel(time = slice('1979-01-01','2014-12-30'))
        ds['time'] = pandas_time

        ds_tropical_mean = metcalc.xr_fldmean(ds.sel(lat=slice(-lat_bound,lat_bound)))
        ds_tm_ym = ds_tropical_mean.resample(time='Y').mean()

        model_list.append(model)
        ds_tm_ym_list.append(ds_tm_ym)
        ds_list.append(ds)


    mod_id_ar = xr.DataArray(model_list[:], name='model_id',dims='model_id')
    
    ds_tropical_mean = xr.concat(ds_tm_ym_list, mod_id_ar)
    
    return ds_list, ds_tropical_mean
```

```python
ds_ta_list, ds_ta_tm = format_data(dset_dict_ta, is_3D = True)
```

```python
ds_ts_list, ds_ts_tm = format_data(dset_dict_ts_pr, is_3D = False)
```

## Process data

```python

def format_data_piControl(dset_dict, is_3D,lat_bound=20) :
    model_list = []
    ds_tm_list=[]
    ds_tm_tm_list=[]
    ds_list = []
    levels=[100000, 85000, 70000, 60000, 50000, 30000, 25000, 20000, 15000, 10000, 5000]
    all_levels = [100000., 92500., 85000., 70000., 60000., 50000., 40000., 30000., 25000.,
                20000.,  15000.,  10000., 7000., 5000., 3000., 2000., 1000., 500., 100.]

    for key, value in dset_dict.items():

        ds = dset_dict[key].squeeze()
        model = ds.source_id
        
        if is_3D == True :
            #for some models (ACCESS) the plev coordinate is weird. Therefore:
            ds['plev'] = all_levels
            ds = ds.sel(plev=levels)
        print("\n" + model )
        if 'latitude' in ds :
            ds = ds.rename( {'latitude' : 'lat'})
            ds = ds.rename( {'longitude' : 'lon'})
        
        ds_tm = metcalc.xr_fldmean(ds.sel(lat=slice(-lat_bound,lat_bound)))
        ds_tm_tm = ds_tm.mean('time')
        model_list.append(model)
        ds_tm_list.append(ds_tm)
        ds_tm_tm_list.append(ds_tm_tm)
        ds_list.append(ds)


    mod_id_ar = xr.DataArray(model_list[:], name='model_id',dims='model_id')
    
    ds_tm_tm = xr.concat(ds_tm_tm_list, mod_id_ar)
    
    return ds_list, ds_tm_list, ds_tm_tm
```

```python
def format_2(ds,is_3D,lat_bound=20) :
    model_list = []
    ds_tm_list=[]
    ds_tm_tm_list=[]
    ds_list = []
    levels=[100000, 85000, 70000, 60000, 50000, 30000, 25000, 20000, 15000, 10000, 5000]
    all_levels = [100000., 92500., 85000., 70000., 60000., 50000., 40000., 30000., 25000.,
                20000.,  15000.,  10000., 7000., 5000., 3000., 2000., 1000., 500., 100.]

    model = ds.source_id
    if is_3D == True :
        ds['plev'] = all_levels
        ds = ds.sel(plev=levels)
    print("\n" + model )
    if 'latitude' in ds :
        ds = ds.rename( {'latitude' : 'lat'})
        ds = ds.rename( {'longitude' : 'lon'})

    ds_tm = metcalc.xr_fldmean(ds.sel(lat=slice(-lat_bound,lat_bound)))
    ds_tm_tm = ds_tm.mean('time')

    return ds, ds_tm, ds_tm_tm
```
