```python
import xarray as xr
import numpy as np
```

Downloaded here: http://web.science.unsw.edu.au/~stevensherwood/radproj/index.html


```python
iuk=xr.open_dataset("../data/radiosondes/T_monthly_v2.2015.nc")
```

```python
iuk
```

```python
lat_border=20

iuk_=iuk.set_coords(['idlist','levels','month_dates'])
iuk__=iuk_.rename({'levels': 'lev', 'month_dates' : 'month', 'idlist' : 'sta', 'monthly_means' : 'T_mm'})
print(iuk__)
tropics_ = iuk__.where(iuk__.latlist > - lat_border,drop=True)
tropics = tropics_.where(tropics_.latlist < lat_border,drop=True)
tropics['T_mm'] = tropics.T_mm.where( tropics.T_mm > -9999., np.nan)
tropics
```

```python

lat=tropics['latlist']
lon=tropics['lonlist']

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

fig = plt.figure(figsize=(18, 3))

ax = fig.add_subplot(1, 1, 1,projection=ccrs.PlateCarree())
ax.set_extent([0, 359, -lat_border, lat_border], crs=ccrs.PlateCarree())

ax.scatter(lon,lat,c='red',transform=ccrs.PlateCarree())

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)

```

Save file:

```python
tropics = tropics.rename({ 'sta' : 'station', 'lev' : 'plev'})
#make int to float:
tropics['plev']=tropics.plev.values/1.

tropics.to_netcdf(path='../data/radiosondes/iuk_'+ str(lat_border) +'N-' + str(lat_border) + 'S.nc' , mode='w')
```
